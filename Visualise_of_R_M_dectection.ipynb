```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

# Load data
df = pd.read_csv("/content/sonar.all-data.csv")

X = df.drop(columns=["Label"])
y = df["Label"].map({"R":0, "M":1})  # FIXED: convert to 0/1

# Step 1: PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 2: Logistic Regression
model = LogisticRegression()
model.fit(X_pca, y)

# Step 3: Decision Boundary
x_min, x_max = X_pca[:,0].min() - 1, X_pca[:,0].max() + 1
y_min, y_max = X_pca[:,1].min() - 1, X_pca[:,1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.2, cmap="coolwarm")
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="coolwarm", edgecolor="k")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("Logistic Regression Decision Boundary (Rock vs Mine)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score

# Train logistic regression directly on original features
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:,1]

fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0,1],[0,1],"--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Logistic Regression)")
plt.legend()
plt.show()
```
